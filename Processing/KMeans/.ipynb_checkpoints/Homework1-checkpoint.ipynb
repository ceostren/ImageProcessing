PART 1 - Sharpen a blurred colored image using a Gaussian filter

```python
import numpy as np
import skimage
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
import os
import scipy.misc as sm
from scipy.ndimage.filters import convolve

#Read and display the image to analyze
def load_img(filename):
    if os.path.isfile(os.getcwd() + '/' + filename):
        img = mpimg.imread(filename)
    return img

#Create a Gaussian filter of n x n size and width sigma
def gaussian_filter(n, sigma=1):
    n = int(n) // 2
    x, y = np.mgrid[-n:n+1, -n:n+1]
    g =  (np.exp(-((x**2 + y**2) / (2.0*sigma**2)))) / (2.0 * np.pi * sigma**2)
    return g

#draw a circular patch on a color image Img; the patch is centered at r,c, with radius rad
#and color col (0-black; 1-red; 2-cyan; 3-green)
#Note that this function writes directly into Img. 
def draw_patch(Img, r, c, rad=7,col=1):   
    if col == 2:
        p = [0,1.0,1.0] #cyan
    else:
        p = [0,1.0,0.0] #green
    if col == 1: 
        p = [1.0,0.0,0.0]#red
    if col == 0: 
        p = [0.0,0.0,0.0]#black
        
    rr, cc = skimage.draw.circle(r, c, rad)
    Img[rr,cc, 0:3]= p       
 
        
```

```python
img = load_img('timnit_blur.jpg')
g = gaussian_filter(49,1.5)
plt.imshow(img)
plt.show()

#Now convert the image to Lab color scale
lab = skimage.color.rgb2lab(img)

# Now extract only the first channel and perform convolutions on this
L = lab[:,:,0]
smoothed = convolve(L, g) 
#print(smoothed.dtype, smoothed.shape)
newL = 1.8*L - 1.5*smoothed
#newL = (100*(newL - newL.min()) / (newL.max() - newL.min()))
newL = (100* newL  / newL.max())

lab2 = np.zeros(img.shape)
lab2[:,:,0] = newL
lab2[:,:,1] = lab[:,:,1]
lab2[:,:,2] = lab[:,:,2]

img2 = skimage.color.lab2rgb(lab2)
mpimg.imsave('timnit_sharper.jpg', img2)
plt.imshow(img2)
plt.show()


```

```python
img1 = load_img('waldo_onIce.png')
temp1 = load_img('waldo_template.png')

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
ax = axes.ravel()
ax[0].imshow(img1)
#ax[0].imshow(img, cmap="gray")
ax[1].imshow(temp1)
fig.tight_layout()
plt.show()

img = skimage.color.rgb2gray(img1)
temp = skimage.color.rgb2gray(temp1)
```

```python
def crosscorr(Img, f):
    M, N = Img.shape
    m,n = f.shape
    Z = np.zeros((M,N))
    z = np.zeros((m,n))
    maxval = 0
    
    for i in range(1,M-m):
        for j in range(1,N-n):
            z = Img[i:i+m,j:j+n]
            val = np.mean(z * f)
            Z[i,j] = val
            if val > maxval:
                maxval = val
                row,col = int(i+m/2),int(j+n/2)
    
    plt.imshow(Z,cmap="gray",aspect='auto')
    plt.show()
    print(np.max(Z))
    print(row,col)
    
    #Draw a patch where Waldo was found
    draw_patch(img1, row, col,rad=10,col=2)
    plt.imshow(img1,aspect='auto')
    plt.show()
 
    
crosscorr(img,temp)
```

```python
def normcrosscorr(Img, f):
    M, N = Img.shape
    m,n = f.shape
    Z = np.zeros((M,N))
    z = np.zeros((m,n))
    maxval = 0
    
    for i in range(1,M-m):
        for j in range(1,N-n):
            z = Img[i:i+m,j:j+n]
            z_bar = np.mean(z)
            f_bar = np.mean(f)
            f_hat = (f-f_bar)/(0.00001+np.sqrt(np.sum((f-f_bar)**2)))
            z_hat = (z-z_bar)/(0.00001+np.sqrt(np.sum((z-z_bar)**2)))
            val = np.mean(z_hat * f_hat)
            Z[i,j] = val
            if val > maxval:
                maxval = val
                #this is tricky; window is matched a the top left corner
                row,col = int(i+m/2),int(j+n/2)
    
    plt.imshow(Z,cmap="gray",aspect='auto')
    plt.show()
    print(np.max(Z))
    print(row,col)
    
    #Draw a patch where Waldo was found
    draw_patch(img1, row, col, col=3,rad=10)
    plt.imshow(img1,aspect='auto')
    plt.show()
 
    
normcrosscorr(img,temp)
```

```python
def SSD(Img, f):
    M, N = Img.shape
    m,n = f.shape
    Z = np.zeros((M,N))
    z = np.zeros((m,n))
    minval = 10000
    
    for i in range(1,M-m):
        for j in range(1,N-n):
            z = Img[i:i+m,j:j+n]
            val = np.sum((z - f)**2)
            Z[i,j] = val
            if val < minval:
                minval = val
                #this is tricky; window is matched a the top left corner
                row,col = int(i+m/2),int(j+n/2)
    
    plt.imshow(Z,cmap="gray",aspect='auto')
    plt.show()
    print(np.max(Z))
    print(row,col)
    
    #Draw a patch where Waldo was found
    draw_patch(img1, row, col, col=1,rad=7)
    plt.imshow(img1,aspect='auto')
    plt.show()
 
    
SSD(img,temp)
```
